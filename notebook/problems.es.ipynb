## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import random
import numpy as np

def lanzamientos():
    dado_1 = random.randint(1,6)
    dado_2 = random.randint(1,6)
    return dado_1 + dado_2

iteraciones = 1000
valores_mayores_a_siete = 0

for _ in range(iteraciones):
    suma = lanzamientos()
    if suma > 7 or suma %2 == 0:
        valores_mayores_a_siete +=1

probabilidad = (valores_mayores_a_siete / iteraciones)*100

print(f"La probabilidad es del:",probabilidad,"%")
    
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO

# Parámetros del experimento
repeticiones = 1000
numero_de_bolas_a_sacar = 5

# Contadores para los resultados
contador_bolas_blancas_y_rojas = 0
contador_bolas_del_mismo_color = 0

# Simular el experimento 1000 veces
for _ in range(repeticiones):
    # Extraer 5 bolas con reemplazo usando los valores del diccionario
    sacadas = [ball_box[random.randint(0, 59)] for _ in range(numero_de_bolas_a_sacar)]
    
    # Contar cuántas bolas de cada color fueron extraídas
    blancas = sacadas.count('White')
    rojas = sacadas.count('Red')
    verdes = sacadas.count('Green')
# Caso 1: 3 blancas y 2 rojas
    if blancas == 3 and rojas == 2:
        contador_bolas_blancas_y_rojas += 1
    
    # Caso 2: Todas del mismo color
    if blancas == 5 or rojas == 5 or verdes == 5:
        contador_bolas_del_mismo_color += 1

# Calcular las probabilidades
tomar_3_bolas_blancas_y_2_rojas = (contador_bolas_blancas_y_rojas / repeticiones) * 100
tomar_bolas_del_mismo_color = (contador_bolas_del_mismo_color / repeticiones)* 100

# Imprimir los resultados
print(f'La probabilidad de sacar 3 bolas blancas y 2 bolas rojas es del:', tomar_3_bolas_blancas_y_2_rojas,'%')
print(f'La Probabilidad de sacar las bolas del mismo color es del:', tomar_bolas_del_mismo_color,'%')
```
